```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

```python
test_acc_Heterog_train_Heterog_init = np.load('test_acc_Heterog_train_Heterog_init.npy')
test_acc_Homog_train_Heterog_init = np.load('test_acc_Homog_train_Heterog_init.npy')
test_acc_Homog_train_Homog_init = np.load('test_acc_Homog_train_Homog_init.npy')
test_acc_Heterog_train_Homog_init = np.load('test_acc_Heterog_train_Homog_init.npy')
```

```python
plt.plot(test_acc_Homog_train_Homog_init[0],test_acc_Homog_train_Homog_init[1],label='homog_train_homog_init')
plt.plot(test_acc_Homog_train_Heterog_init[0],test_acc_Homog_train_Heterog_init[1], label='homog_train_heterog_init')
plt.plot(test_acc_Heterog_train_Homog_init[0],test_acc_Heterog_train_Homog_init[1],label='heterog_train_homog_init')
plt.plot(test_acc_Heterog_train_Heterog_init[0],test_acc_Heterog_train_Heterog_init[1], label='heterog_train_heterog_init')
plt.xlabel("Epoch")
plt.ylabel("Test accuracy")
plt.legend()
plt.show()
```

```python
beta_time_constant_distributions_Heterog_train_Heterog_init =   np.load('beta_time_constant_distributions_Heterog_train_Heterog_init.npy')
beta_time_constant_distributions_Homog_train_Heterog_init   =   np.load('beta_time_constant_distributions_Homog_train_Heterog_init.npy')
beta_time_constant_distributions_Heterog_train_Homog_init   =   np.load('beta_time_constant_distributions_Heterog_train_Homog_init.npy')
beta_time_constant_distributions_Homog_train_Homog_init     =   np.load('beta_time_constant_distributions_Homog_train_Homog_init.npy')
```

```python
plt.figure(figsize=(9,6))
ax1 = plt.subplot(2,2,4)
ax2 = plt.subplot(2,2,3, sharex=ax1)
ax3 = plt.subplot(2,2,2, sharex=ax1)
ax4 = plt.subplot(2,2,1, sharex=ax1)
ax1.hist(-0.5/np.log(beta_time_constant_distributions_Heterog_train_Heterog_init), bins = 30, range = [0,100], label='Heterog_train_Heterog_init', color= 'r')
ax2.hist(-0.5/np.log(beta_time_constant_distributions_Homog_train_Heterog_init), bins = 30, range = [0,100], label='Homog_train_Heterog_init', color= 'orange')
ax3.hist(-0.5/np.log(beta_time_constant_distributions_Heterog_train_Homog_init), bins = 30, range = [0,100], label='Heterog_train_Homog_init', color= 'green')
ax4.hist(-0.5/np.log(beta_time_constant_distributions_Homog_train_Homog_init), bins = 30, range = [0,100], label='Homog_train_Homog_init')
ax1.legend()
ax2.legend()
ax3.legend()
ax4.legend()
plt.xlabel('Membrane time constant (ms)')
plt.show()
```
